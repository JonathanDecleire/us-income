```python
# linear algebra
import numpy as np

# data processing
import pandas as pd

# data visualization
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
train_df = pd.read_csv('C:/Users/ACER/Desktop/pull github repo/LIE-Thomas-1.26/content/additional_resources/datasets/US Income/cleaned/data_train.csv')

df = train_df.copy()
```

```python
def income_count():
    plt.title('Number of people in each category of income')
    sns.countplot(x='income', data=df)
```

```python
def age_count():
    sns.displot(df, x="age", bins=20)
    plt.title('Number of people in each age interval')
```

```python
def education_count():
    sns.countplot(data=df, y='education', order=df['education'].value_counts().index)
    plt.title('Number of people in each education category ranked')
```

```python
def education_num_count():
    sns.countplot(data=df, x='education-num')
    plt.title('Number of people categorized by number of years of education')
```

```python
def income_age_count():
    df.groupby('income').age.hist(bins=180)
    plt.title('Sample population grouped by income category and distributed by age')
```
